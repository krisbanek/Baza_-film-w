# Zwróć listę 10 najwyżej ocenianych filmów (vote_average), których liczba głosów (vote_count) jest większa od 3. kwartyla rozkładu liczby głosów.

```python
df[df['vote_count'] > df['vote_count'].quantile(0.75)].sort_values(by='vote_average', ascending=False).iloc[:10]
```

# Średni przychód oraz budżet w latach 2010 - 2016

```python
k = df[['original_title','revenue','budget','release_date']][(df['release_date'] >= '2010-01-01') & (df['release_date'] <= '2016-12-31')]
```

```python
k['release_date'] = pd.to_datetime(df['release_date'])
k['release_date'] = k['release_date'].dt.strftime('%Y')
```

```python
graph_data = k.groupby(pd.Grouper(key='release_date')).mean().sort_values(by='release_date')
```

```python
def million(x, pos):
    return '{:2.1f} M'.format(x/1000000)
```

```python
graph_data['revenue M'] = graph_data[('revenue')]/1000000
graph_data['budget M'] = graph_data[('budget')]/1000000
```

```python
fig, ax = plt.subplots()

#fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
years = graph_data.index
revenue = graph_data['revenue M']
budget = graph_data['budget M']

axes.bar(years,revenue,label='revenue M')
axes.plot(budget,label='budget M',color='red', linewidth=3)

axes.legend(loc=(1.05,0.0))
axes.set_title('Średni przychód oraz budżet w latach 2010 - 2016')
```

# Połączenie bazy filmów z bazą gatunków, tak aby w bazie filmów można było odczytać nazwę gatunku filmu.

```python
df_movies = pd.read_csv('tmdb_movies.csv')
df_genre = pd.read_csv('tmdb_genres.csv')
```

```python
df_genre.rename(columns={'Unnamed: 0':'genre_id'},inplace=True)
df_genre.keys()
```

```python
pd.merge(df_movies,df_genre,how='inner',on='genre_id')
```

# Jaki gatunek filmu z bazy pojawia się w niej najczęściej? Ile filmów tego gatunku znajduje się w bazie?

```python
df_mg = pd.merge(df_movies,df_genre,how='inner',on='genre_id')
```

```python
D = df_mg['genres'].value_counts().index[0]
E = df_mg['genres'].value_counts()[0]
```

```python
print("Gatunek "+ str(D) + " pojawia się w niej najczęściej w bazie.")
print(str(E) + " filmów tego gatunku znajduje się w bazie.")
```

# Filmy, którego gatunku trwają średnio najdłużej (runtime)?

```python
F = df_mg.groupby(pd.Grouper(key='genres')).agg({'runtime':'mean'}).sort_values(by='runtime',ascending=False).index[0]
```

```python
print("Filmy gatunku " + str(F) + " trwają średnio najdłużej.")
```
